```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,4,9,16,25]
#create a line plot
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Basic line plot")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
##create a customized line plot
plt.plot(x,y,color="red",linestyle="dotted",marker="o",linewidth=5,markersize=10)
plt.grid(True)
```

```python
##multiple plots
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figuresize=(9,5)
plt.subplot(1,2,1)
plt.plot(x,y1,color="green")
plt.title("plot 1")


plt.subplot(1,2,2)
plt.plot(y1,y2,color="yellow")
plt.title("plot 2")

plt.subplot(2,2,3,)
plt.plot(x,y2,color="blue")
plt.title("plot 3")
plt.tight_layout(pad=0.3)

plt.show()

```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figuresize=(9,5)
plt.subplot(1,2,1)
plt.plot(x,y1,color="green")
plt.title("plot 1")

plt.subplot(1,2,2)
plt.plot(y1,y2,color="yellow")
plt.title("plot 2")

plt.subplot(2,2,3)
plt.plot(x,y2,color="blue")
plt.title("plot 3")
```

```python
##bar plot
categories=["A","B","C"]
values=[2,3,4]
plt.bar(categories,values)
```

```python
##scatter plot
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
plt.scatter(x,y,color="blue",marker="x")
```

```python
#piechart
labels=['A','B','C']
sizes=[20,40,40]
colors=["red","green","blue"]
explode=[0.2,0,0]

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
```

```python
import pandas as pd
sales_data=pd.read_csv("sales_data.csv")
sales_data.head()
```

```python
sales_data.info()
```

```python
#plot by product
total_sales_product=sales_data.groupby("Item Type")['Total Revenue'].sum()
print(total_sales_product)
```

```python
total_sales_product.plot(kind="bar", color="teal")
```

```python
#plot sales trend over time
sales_trend=sales_data.groupby("Order Date")["Total Revenue"].sum().reset_index()
plt.plot(sales_trend['Order Date'],sales_trend['Total Revenue'])
```
