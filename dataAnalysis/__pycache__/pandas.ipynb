```python
import pandas as pd
df=pd.read_csv('country_full.csv')
df.head(5)
df.tail(10)
df.describe()
```

```python
df.dtypes
```

```python
#handling missing values
df.isnull()
```

```python
df.isnull().any()#by default the axis is 0
```

```python
df.isnull().sum()
```

```python
df.fillna("null")#to replace the missing value cell
```

```python
#filling missing value with the mean of the colums
df['intermediate-region']=df['country-code'].fillna((df['country-code'].mean()))
df
```

```python
df.dtypes
```

```python
#changing the datatypes
df=df.rename(columns={'name':'country_name'})
df.head()
```

```python
df['new_region_code']=df["region-code"].fillna(df['region-code'].mean()).astype(int)#filled all NaN values with the mean and changed the datatype to int form float
df
```

```python
df['new_region_code']=df['region-code'].apply(lambda x:x*2)
df.head()
```

```python
#data aggregating and grouping
df.head()
grouped_mean=df.groupby('country_name')['region-code'].mean();
print(grouped_mean)
```

```python
#merging and joining dataframes
df1=pd.DataFrame({'key':['A','B','C'], 'value1':[1,2,3]})
df2=pd.DataFrame({'key':['A','B','D'],'value2':[4,5,6]})
#merging data frames on the "key column"
pd.merge(df1,df2, on="key",how="inner")
```

```python
pd.merge(df1,df2,on="key",how="outer")
```

```python
import pandas as pd
df1=pd.DataFrame({'key':['A','B','C'], 'value1':[1,2,3]})
df2=pd.DataFrame({'key':['A','B','D'],'value2':[4,5,6]})
#merging data frames on the "key column"
pd.merge(df1,df2, on="key",how="inner")
pd.merge(df1,df2,on="key",how="left")
```

```python
pd.merge(df1,df2,on="key",how="right")
```
